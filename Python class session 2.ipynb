Lets create two variables with different integers and make a small script stating the biggest number

````python
x = 98
y= 29
````

````python
if x > y:
    larger_number = x
else:
    larger_number = y

# Print the result
print("The larger number is:", larger_number)
````

Do you remember how to make the user input a value?

````python
x = int(input("Please enter the first number: "))
y = int(input("Please enter the second number: "))
if x > y:
    larger_number = x
else:
    larger_number = y

# Print the result
print("The larger number is:", larger_number)
````
